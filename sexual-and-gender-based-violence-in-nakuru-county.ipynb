**Visualize the data**

**Visualize correlation plot of the whole dataset**

```python
sns.heatmap(df.corr())
```

**Do a barplot of data for the number of Perpetrators in relation to date and Month**

```python
sns.barplot(data=df, x="No_of_perpetrators", y="Year/Month")
```

**Do a bar plot for the number of rape survivors within each month**

```python
sns.barplot(data=df,x ='No_of_rape_survivors',y='Year/Month' )
```

**Do a bar plot of Total survivors with disability**

```python
sns.barplot(data=df, x="Total_survivors_with_disability", y="Year/Month")
```

**Plot Jointplot with all the data**

```python
sns.jointplot(data=df, x="No_of_perpetrators", y="No_of_rape_survivors")
```

**Joint plot with the hist kind**

```python
sns.jointplot(data=df, x="No_of_perpetrators", y="No_of_rape_survivors",kind="hist")
```

**Histplot of the data Number of perpetrators**

```python
sns.histplot(data=df, y="No_of_perpetrators")
```

**Flip the histogram above**

```python
sns.histplot(data=df, x="No_of_perpetrators")
```

**Add bins to the histogram**

```python
sns.histplot(data=df, x="No_of_perpetrators", bins=20)
```

**Do a histogram of Number of rape survivors within every month**

```python
sns.histplot(data=df, x="No_of_rape_survivors",y="No_of_perpetrators", hue="Year/Month")
```

**Do a bar chart for each month**

```python
df.No_of_rape_survivors.value_counts().nlargest(10).plot(kind='bar',figsize=(15,5))
plt.title("No of Rape Survivors by Month")
plt.ylabel("Year/Month")
plt.xlabel("No of  Rape Survivors")
```
